# Prediction

```python
X = df.drop(columns = ['price'], axis  = 1)
y = df['price']
```

```python
X.head()
```

```python
X_onehot_encode = pd.get_dummies(X, columns = ['model', 'transmission', 'fuelType'], drop_first = True)
```

```python
X_onehot_encode
```

```python
X_onehot_encode = X_onehot_encode.astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()

columns = ['model', 'transmission', 'fuelType']

X_label = X


for i in columns:
    X_label[i] = encode.fit_transform(X_label[i])
```

```python
X_label
```

```python
X['model'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

numeric_cols = ['year', 'mileage', 'tax', 'mpg']

scaler = StandardScaler()

X_onehot_encode[numeric_cols] = scaler.fit_transform(X_onehot_encode[numeric_cols])
```

```python
X_onehot_encode
```

```python
X_label.columns
```

```python
X_label[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(X_label[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
X_label
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_onehot_encode, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))
print("Adjusted r2: ",adjusted_r2 )
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_label, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test, y_pred)
r2
```
